## Introduction

This notebook presents a self-directed comparative study of supervised classification models using the Iris dataset. The focus is on hyperparameter sensitivity and model behavior rather than optimizing benchmark performance. Implemented models include Support Vector Machines, Logistic Regression, Random Forests, k-Nearest Neighbors, and Neural Networks.

Decision boundaries are analyzed using two feature subsets: petal measurements, which are strongly separable, and sepal measurements, which exhibit greater class overlap. ## The Iris Dataset

The Iris dataset is a classic multiclass classification dataset originally introduced by Ronald Fisher. It contains 150 observations across three iris species, with four continuous features: sepal length, sepal width, petal length, and petal width. The dataset is widely used for illustrating supervised learning, decision boundaries, and model comparison due to its low dimensionality and partially separable structure.

Data source: Kaggle
Original dataset archive: UCI Machine Learning Repository

# Exploratory Data Analysis

```python
iris = pd.read_csv('data/iris.csv') # Load the Iris dataset from a CSV file
print(iris.head()) # Display the first few rows of the dataset to verify it loaded correctly
```

```python
print("Missing values:")
print(iris.isnull().sum())
print()
print("Duplicate rows:", iris.duplicated().sum())
iris.dropna(inplace=True)
iris.drop_duplicates(inplace=True)
print("Dropped missing values and duplicates.")
```

```python
iris = iris.drop(columns=['Id']) # Drop the 'Id' column as it is not useful for analysis
```

```python
iris.info() # Display information about the dataset, including data types and non-null counts
```

```python
iris.describe() # Get summary statistics for the numerical columns in the dataset
```

```python
iris_markers = ['o', 's', 'D']
iris_colors = ['blue', 'orange', 'green']

setosa = iris[iris['Species'] == 'Iris-setosa']
versicolor = iris[iris['Species'] == 'Iris-versicolor']
virginica = iris[iris['Species'] == 'Iris-virginica']
```

### Sepal

```python
plt.scatter(setosa['SepalLengthCm'], setosa['SepalWidthCm'], 
            color='blue', marker='o', label='Setosa')

plt.scatter(versicolor['SepalLengthCm'], versicolor['SepalWidthCm'], 
            color='orange', marker='s', label='Versicolor')

plt.scatter(virginica['SepalLengthCm'], virginica['SepalWidthCm'], 
            color='green', marker='D', 
            label='Virginica')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.show()
```

Using only sepal length and sepal width, the Setosa class appears clearly separated from the other two species. In contrast, Versicolor and Virginica exhibit substantial overlap in this feature space. This overlap suggests that these two classes are not linearly separable when restricted to sepal measurements alone, indicating a limitation of these features for distinguishing between them.

## Petal

```python
plt.scatter(setosa['PetalLengthCm'], setosa['PetalWidthCm'], 
            color='blue', marker='o', label='Setosa')

plt.scatter(versicolor['PetalLengthCm'], versicolor['PetalWidthCm'], 
            color='orange', marker='s', label='Versicolor')

plt.scatter(virginica['PetalLengthCm'], virginica['PetalWidthCm'], 
            color='green', marker='D', label='Virginica')

plt.xlabel('Petal Length (cm)') 
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width')
plt.legend()
plt.show()
```

Using only petal length and petal width, the class separation is substantially clearer. Setosa remains completely separated from the other species, and the overlap between Versicolor and Virginica is minimal compared to the sepal feature space. This indicates that petal measurements provide stronger discriminatory information for classification than sepal measurements alone.

# Test Train Split

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
iris.shape
```

```python
train, test = train_test_split(
    iris, test_size=0.2, random_state=42, stratify=iris["Species"]
)

X_sepal_train = train[["SepalLengthCm", "SepalWidthCm"]].values
X_sepal_test  = test[["SepalLengthCm", "SepalWidthCm"]].values

X_petal_train = train[["PetalLengthCm", "PetalWidthCm"]].values
X_petal_test  = test[["PetalLengthCm", "PetalWidthCm"]].values

y_train = train["Species"].values
y_test  = test["Species"].values

scaler_sepal = StandardScaler()
X_sepal_train_std = scaler_sepal.fit_transform(X_sepal_train)
X_sepal_test_std  = scaler_sepal.transform(X_sepal_test)

scaler_petal = StandardScaler()
X_petal_train_std = scaler_petal.fit_transform(X_petal_train)
X_petal_test_std  = scaler_petal.transform(X_petal_test)

print("Shape of X_sepal_train_std:", X_sepal_train_std.shape)
print("Shape of X_petal_train_std:", X_petal_train_std.shape)
print("Shape of X_sepal_test_std:", X_sepal_test_std.shape)
print("Shape of X_petal_test_std:", X_petal_test_std.shape) 
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
X_train_combined = np.hstack((X_sepal_train_std, X_petal_train_std))
X_test_combined = np.hstack((X_sepal_test_std, X_petal_test_std))

np.savez(
    "data/iris_splits.npz",
    X_sepal_train_std=X_sepal_train_std,
    X_sepal_test_std=X_sepal_test_std,
    X_petal_train_std=X_petal_train_std,
    X_petal_test_std=X_petal_test_std,
    X_train_combined=X_train_combined,
    X_test_combined=X_test_combined,
    y_train=y_train,
    y_test=y_test
)
```
